#### Import the Regural Expression (re) library which is mainly used in text preprocessing

```python
import re
```

```python
# we remove the extra characters other than a to z and A to Z
re.sub('[^a-zA-Z]',' ', data)
```

```python
#we try to lower case them.
data=data.lower()
data
```

```python
list = data.split()
list
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
ps=PorterStemmer()
```

```python
#we use the stemming technique to find the root or base words then we remove stopwords that do not
#add significant value to our processing
```

```python
review=[]
for word in list:
    if word not in set(stopwords.words('english')):
        review.append(ps.stem(word))
review
```

```python
' '.join(review)
```

```python
### Now I do same task for each & every row
```

```python
corpus=[]
sentences=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ', messages['title'][i])
    review=review.lower()
    list=review.split()
    review=[ps.stem(word) for word in list if word not in set(stopwords.words('english'))]
    sentences=' '.join(review)
    corpus.append(sentences)
```

```python
corpus[0]
```

```python
corpus
```

```python
len(corpus)
```

```python
#Now we use bag of words model, to convert text into matrix of numbers
```

```python
## Applying Countvectorizer
# Creating the Bag of Words model

from sklearn.feature_extraction.text import CountVectorizer
```

```python
## max_features=5000, it means I just need top 5000 features 
#eso in ngram,i have Given (1,3),so it will take the combination of 1 word,then 2 words 
#then 3 words

cv=CountVectorizer(max_features=5000,ngram_range=(1,3))
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
X.shape
#ie we get 5000 features now
```

```python
X
```

```python
cv.get_feature_names()[0:20]
```

```python
messages.columns
```

```python
y=messages['label']
```

```python
## Divide the dataset into Train and Test
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
X_test
```

```python
X_test.shape
```

```python
X_train
```

###  MultinomialNB Algo

```python
#this algo works well with text data

from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
```

```python
classifier.fit(X_train,y_train)
```

```python
pred=classifier.predict(X_test)
pred
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,pred)
```

```python
cm=metrics.confusion_matrix(y_test,pred)
cm
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### make your confusion amtrix more user-friendly

plt.imshow(cm,interpolation='nearest',cmap=plt.cm.Blues)
plt.colorbar()
plt.title('Confusion Matrix')
labels=['positive','negative']
tick_marks=np.arange(len(labels))
plt.xticks(tick_marks,labels)
plt.yticks(tick_marks,labels)
```

#### Passive Aggressive Classifier Algorithm

```python
linear_clf=PassiveAggressiveClassifier()
```

```python
cm2=metrics.confusion_matrix(y_test,predictions)
cm2
```

```python
plt.imshow(cm2,interpolation='nearest',cmap=plt.cm.Blues)
plt.colorbar()
plt.title('Confusion Matrix')
labels=['positive','negative']
tick_marks=np.arange(len(labels))
plt.xticks(tick_marks,labels)
plt.yticks(tick_marks,labels)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, cohen_kappa_score, f1_score, recall_score, precision_score

# Convert positive class label to string
pos_label = '1'

# Compute performance metrics for Passive Aggressive Classifier
pac_confusion_matrix = confusion_matrix(y_test, predictions)
pac_accuracy = accuracy_score(y_test, predictions)
pac_kappa = cohen_kappa_score(y_test, predictions)
pac_precision = precision_score(y_test, predictions, pos_label=pos_label)
pac_recall = recall_score(y_test, predictions, pos_label=pos_label)
pac_f1 = f1_score(y_test, predictions, pos_label=pos_label)

# Compute performance metrics for Multinomial Naive Bayes
mnb_confusion_matrix = confusion_matrix(y_test, pred)
mnb_accuracy = accuracy_score(y_test, pred)
mnb_kappa = cohen_kappa_score(y_test, pred)
mnb_precision = precision_score(y_test, pred, pos_label=pos_label)
mnb_recall = recall_score(y_test, pred, pos_label=pos_label)
mnb_f1 = f1_score(y_test, pred, pos_label=pos_label)

# Print the performance metrics
print("Passive Aggressive Classifier:")
print("Confusion Matrix:\n", pac_confusion_matrix)
print("Accuracy:", pac_accuracy)
print("Kappa:", pac_kappa)
print("Precision:", pac_precision)
print("Recall Score:", pac_recall)
print("F1 Score:", pac_f1)

print("\nMultinomial Naive Bayes:")
print("Confusion Matrix:\n", mnb_confusion_matrix)
print("Accuracy:", mnb_accuracy)
print("Kappa:", mnb_kappa)
print("Precision:", mnb_precision)
print("Recall Score:", mnb_recall)
print("F1 Score:", mnb_f1)
```
